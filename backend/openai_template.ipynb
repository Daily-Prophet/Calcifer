````python
import openai
import os
from dotenv import find_dotenv, load_dotenv
_ = load_dotenv(find_dotenv())

openai.api_base = os.environ["OPENAI_API_BASE"]
openai.api_version = os.environ["OPENAI_API_VERSION"]
openai.api_key = os.environ["OPENAI_API_KEY"]
openai.api_type = os.environ["OPENAI_API_TYPE"]
````

````python
def get_completion(prompt, model="gpt-3.5-turbo", log=False):
    # we should follow the openai chat api interface messages order to build chat context: system, user, assistant, user, assistant, user...
    # if not provide the first system message, the api will use the default system message.
    # FYI. https://platform.openai.com/docs/api-reference/chat/create
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        deployment_id="gpt35-16k", model=model, messages=messages, temperature=0
    )
    if log:
        print(response)
    print(response.choices[0].message.content)
    return response.choices[0].message.content
````

## 1.3 模型局限：幻觉
- 即使训练数据包含了大量的知识，但是它并未完全记住它所看到的信息，因此它并不非常了解它的知识边界（我个人认为是习得了某种模式）。

- 处理办法：要求模型首先从文本中找到任何相关引用，然后要求它使用这些引用来回答问题。
